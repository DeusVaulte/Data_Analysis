```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Load the dataset
file_path = 'ecommerce_dataset.csv'  # Replace with your file path
data = pd.read_csv(file_path)


label_encoder = LabelEncoder()
data['Warehouse_block'] = label_encoder.fit_transform(data['Warehouse_block'])


# 1. Data Summary
print("Data Summary:")
print(data.info())
print(data.describe())

# 2. Check for Missing Values
print("\nMissing Values:")
print(data.isnull().sum())

# 3. Data Distribution
# Plot distributions of numerical features
numerical_features = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.histplot(data[feature], kde=True)
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()

# 4. Categorical Data Analysis
# Plot countplots of categorical features
categorical_features = ['Mode_of_Shipment', 'Product_importance', 'Gender']
plt.figure(figsize=(12, 8))
for i, feature in enumerate(categorical_features):
    plt.subplot(2, 2, i+1)
    sns.countplot(data[feature])
    plt.title(f'Count of {feature}')
plt.tight_layout()
plt.show()

# 5. Correlation Analysis
# Compute correlation matrix
corr_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# 6. Target Analysis
# Plot the distribution of the target variable
plt.figure(figsize=(6, 4))
sns.countplot(data['Reached.on.Time_Y.N'])
plt.title('Distribution of Reached.on.Time_Y.N')
plt.show()

# Analyze the relationship between target and other features
plt.figure(figsize=(12, 8))
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.boxplot(x='Reached.on.Time_Y.N', y=feature, data=data)
    plt.title(f'{feature} vs Reached.on.Time_Y.N')
plt.tight_layout()
plt.show()

# Analyze the relationship between categorical features and the target
plt.figure(figsize=(12, 8))
for i, feature in enumerate(categorical_features):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=feature, hue='Reached.on.Time_Y.N', data=data)
    plt.title(f'{feature} vs Reached.on.Time_Y.N')
plt.tight_layout()
plt.show()
```
